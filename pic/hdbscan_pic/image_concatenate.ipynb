```python
import os
from PIL import Image
```

```python
# --- 配置参数 ---
input_folder = 'D:/MachineLearning/STAT_437_Final_Project/pic'  # 存放 90 多张图片的文件夹路径
output_name = 'D:/MachineLearning/STAT_437_Final_Project/pic/combined_result.jpg'   # 输出文件名
images_per_row = 5                   # 每行 5 张

# 获取所有图片文件
img_files = [f for f in os.listdir(input_folder) if f.lower().endswith(('png', 'jpg', 'jpeg'))]
img_files.sort()  # 按文件名排序，确保顺序正确

if not img_files:
    print("文件夹里没找到图片！")
    exit()

# 打开第一张图获取尺寸（假设所有图片尺寸一致）
sample_img = Image.open(os.path.join(input_folder, img_files[0]))
img_w, img_h = sample_img.size

# 计算总行数和画布总尺寸
total_imgs = len(img_files)
rows = (total_imgs + images_per_row - 1) // images_per_row
canvas_w = img_w * images_per_row
canvas_h = img_h * rows

# 创建空白大画布
canvas = Image.new('RGB', (canvas_w, canvas_h), (255, 255, 255))

# 循环拼接
for index, file_name in enumerate(img_files):
    img = Image.open(os.path.join(input_folder, file_name))
    
    # 如果尺寸不一，可以取消下面这行的注释进行缩放
    # img = img.resize((img_w, img_h)) 
    
    row_idx = index // images_per_row
    col_idx = index % images_per_row
    
    x = col_idx * img_w
    y = row_idx * img_h
    canvas.paste(img, (x, y))

# 保存结果
canvas.save(output_name, quality=90)
print(f"拼接完成！已保存为 {output_name}")
```
